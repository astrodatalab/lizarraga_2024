```python

#Some images have been excluded due to too much noise/no appparent features that can be extracted

from IPython.display import display
import pandas as pd

# Path to the CSV file
csv_path = './results.csv'

# Load the CSV file into a DataFrame
results_df = pd.read_csv(csv_path)

# Filter the DataFrame to only include rows where the 'Image File' column starts with 'epoch_15'
filtered_df = results_df[results_df['Image File'].str.startswith('epoch_15')]

# Display the filtered DataFrame
print("Filtered DataFrame:")
display(filtered_df)

# Calculate the average of each numeric column
column_averages = filtered_df.mean(numeric_only=True)

# Print the average for each column
print("\nAverages for each numeric column:")
for column, avg in column_averages.items():
    print(f"{column}: {avg}")
```
