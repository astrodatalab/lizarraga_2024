```python
from IPython.display import display
import pandas as pd

# Path to the CSV file
csv_path = './testing_metrics.csv'

# Load the CSV file into a DataFrame
results_df = pd.read_csv(csv_path)

# Display the DataFrame
display(results_df)

# Calculate the average of each numeric column, excluding 'Sequential Index' and 'Object ID'
column_averages = results_df.drop(columns=['Sequential Index', 'Object ID']).mean(numeric_only=True)

# Print the average for each column
print("\nAverages for each numeric column (excluding 'Sequential Index' and 'Object ID'):")
for column, avg in column_averages.items():
    print(f"{column}: {avg}")
```
