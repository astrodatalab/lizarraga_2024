```python
from IPython.display import display
import pandas as pd

csv_path = './AndrewMetrics.csv'

results_df = pd.read_csv(csv_path)

display(results_df)

column_averages = results_df.drop(columns=['Image Index']).mean(numeric_only=True)

print("\nAverages for each numeric column (excluding 'Image Index'):")
for column, avg in column_averages.items():
    print(f"{column}: {avg}")
```
